```python
import pandas as pd #as = como, gera um apelido (pd é usado por convensão)
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/acoes_pura.xlsx", sheet_name="Principal") #df = DataFrame
#df_principal mostra tudo
#df_principal.head(10) mostra uma quantidade de linhas
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/acoes_pura.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/acoes_pura.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chat_gpt = pd.read_excel("/content/acoes_pura.xlsx", sheet_name="ChatGPT")
df_chat_gpt
```

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy() #importante dar o .copy ao final para apagar o original e substituir pelo novo
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal
```

```python
df_principal['valor_inicial'] = df_principal['valor_final'] / (1+df_principal['Var_pct'])
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left') #VLOOKUP
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código']).copy() #drop deleta a coluna inteira
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format #formata todos os valores do tipo float para duas casas decimais
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Var_pct':'var_pct'}).copy()
df_principal
```

```python
# Aplicando a operação em uma linha
df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável')) #apply é uma funcao que aplica outras funcoes, o lambda faz operacoes de linha a linha
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left') #VLOOKUP
df_principal
```

```python
df_principal = df_principal.drop(columns=['Ticker'])
```

```python
df_principal
```

```python
df_principal = df_principal.merge(df_chat_gpt, left_on='Nome', right_on='Nome da Empresa', how='left') #VLOOKUP
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade em Anos ':'idade'}).copy()
df_principal
```

```python
df_principal.columns
```

```python
df_principal['cat_idade'] = df_principal['idade'].apply(lambda x: "Mais de 100" if x > 100 else ("Menos de 50" if x < 50 else "Entre 50 e 100"))
df_principal
```

```python
# Supondo que df_principal é o seu DataFrame.
# df_principal = pd.DataFrame({
#     'variacao_rs': [...],  # seus valores aqui
#     'resultado': [...]     # 'Subiu' ou 'Desceu'
# })

maior_variacao = df_principal['variacao_rs'].max()
menor_variacao = df_principal['variacao_rs'].min()
media_geral = df_principal['variacao_rs'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao_rs'].mean()

resultado = {
    'Maior': maior_variacao,
    'Menor': menor_variacao,
    'Media': media_geral,
    'Media de quem subiu': media_subiu,
    'Media de quem desceu': media_desceu
}

print(resultado)
```

```python
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index() #groupby agrupar por segmento e somar a variacao_rs, com o comando reset_index ele nao cria um vetor e continua usando o DF
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['variacao_rs'].sum().reset_index() #groupby agrupar por segmento e somar a variacao_rs, com o comando reset_index ele nao cria um vetor e continua usando o DF
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='resultado', y='variacao_rs', text='variacao_rs', title='Variação em Reais por Resultado')
fig.show()
```
