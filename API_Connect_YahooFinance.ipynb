**Aula 02 - Importação**

```python
import pandas as pd
import plotly.express as px
from babel.numbers import format_currency
```

```python
df_principal = pd.read_excel("/content/tabela_acoes_original.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/tabela_acoes_original.xlsx", sheet_name="Total_de_acoes")
df_total_acoes.head(5)
```

```python
df_ticker = pd.read_excel("/content/tabela_acoes_original.xlsx", sheet_name="Ticker")
df_ticker.head(5)
```

```python
df_chat_gpt = pd.read_excel("/content/tabela_acoes_aula.xlsx", sheet_name="Chat_GPT")
df_chat_gpt.head()
```

```python
df_principal = df_principal[["Ativo",	"Data",	"Último (R$)",	"Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)":"valor_final", "Var. Dia (%)":"var_dia_pct"}).copy()
df_principal
```

```python
df_principal["var_pct"] = df_principal['var_dia_pct'] / 100
df_principal["valor_inicial"] = df_principal["valor_final"] / (df_principal["var_pct"] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['variacao'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['resultado'] = df_principal['variacao'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=["Ticker"])
df_principal
```

```python
df_principal = df_principal.merge(df_chat_gpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=["Nome da Empresa"])
df_principal
```

```python
df_principal = df_principal.rename(columns={"Idade (em anos)":"idade"}).copy()
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['idade'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Menor de 50 anos' if x < 50 else 'Entre 50 e 100 anos'))
df_principal
```

```python
maior = df_principal['variacao'].max()
menor = df_principal['variacao'].min()
media = df_principal['variacao'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao'].mean()

# Exibindo os resultados
print("Maior:", 'R$ {:,.2f}'.format(maior))
print("Menor:", 'R$ {:,.2f}'.format(menor))
print("Média:", 'R$ {:,.2f}'.format(media))
print("Média dos que subiram:", 'R$ {:,.2f}'.format(media_subiu))
print("Média dos que desceram:", 'R$ {:,.2f}'.format(media_desceu))
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("Segmento")['variacao'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby("resultado")['variacao'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='resultado', y='variacao', text='variacao', title='Variação R$ por Resultado')
fig.update_traces(texttemplate='%{text}', text=df_analise_saldo['variacao'].apply(lambda x: format_currency(x, 'BRL', locale='pt_BR')))
fig.show()
```

**1. Pesquise com a documentação da biblioteca Plotly ou GPT como mudar a formatação dos números do gráfico de barras**

As alterações de formatos numéricos nessa biblioteca são feitas através do atributo 'tickformat', chamado dentro do atributo 'update_layout', seguido dos eixos, separadamente.

```python
# Fazer o gráfico de pizza no df_análise_segmentos com a mesma biblioteca Plotly;
fig = px.pie(df_analise_segmento, values='variacao', names='Segmento', title='Variação R$ por Resultado')
fig.show()
```

```python
# Fazer o GroupBy da categoria de idades e gerar o gráfico de barras.
df_analise_idades = df_principal.groupby("Cat_idade")['variacao'].sum().reset_index()
df_analise_idades
```

```python
# Gerar o gráfico de barras.
fig = px.bar(df_analise_idades, x='Cat_idade', y='variacao', text='variacao', title='Variação R$ por Categoria de Idade')
fig.update_traces(texttemplate='%{text}', text=df_analise_idades['variacao'].apply(lambda x: format_currency(x, 'BRL', locale='pt_BR')))
fig.show()
```
